```python
with open("clusters.json", "r", encoding="utf-8") as f:
    nkeys = 0 
    nvalues = 0
    data = json.load(f)
    for obj in data:
        nkeys = nkeys + 1
        for key in obj:
            nvalues = nvalues + (len(obj[key]))
    print(nkeys,nvalues)
        
        

```

```python
skill_to_main = {}

for group in data:
    for main_skill, variants in group.items():
        for variant in variants:
            skill_to_main[variant] = main_skill
```

```python
import csv
import ast



updated_rows = []

with open("combine copy.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        # Parse the skills column from string to Python list
        skills = ast.literal_eval(row["skills"])  # Safely turns "['a', 'b']" into ['a', 'b']

        # Replace each skill with the cluster title if applicable
        new_skills = [skill_to_main.get(skill, skill) for skill in skills]

        # Convert back to string for writing into CSV
        row["skills"] = str(new_skills)

        updated_rows.append(row)

# Write to new CSV or overwrite original
with open("output.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=reader.fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
import pandas as pd
import ast
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Load your CSV
df = pd.read_csv("output.csv")

# Convert stringified list of skills to actual lists
df["parsed_skills"] = df["skills"].apply(ast.literal_eval)

# Create graph
G = nx.Graph()

# Add nodes with labels (e.g., job title or company+title)
for idx, row in df.iterrows():
    label = f"{row['company']} - {row['title']}"
    G.add_node(idx, label=label)

# Compare every pair of jobs to find shared skills
for i, j in combinations(df.index, 2):
    skills_i = set(df.loc[i, "parsed_skills"])
    skills_j = set(df.loc[j, "parsed_skills"])
    shared = skills_i & skills_j

    if shared:
        G.add_edge(i, j, weight=len(shared), shared_skills=", ".join(shared))


nx.write_graphml_lxml(G, "my_graph.graphml")
print("graph written")

# Draw graph
pos = nx.spring_layout(G, seed=42)  # force-directed layout

plt.figure(figsize=(12, 10))

# Draw edges with weight-based thickness
weights = [G[u][v]['weight'] for u, v in G.edges()]
nx.draw_networkx_edges(G, pos, width=[0.5 + w for w in weights], alpha=0.6)

# Draw nodes and labels
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=600)
labels = {node: G.nodes[node]["label"] for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=8)

plt.title("Job Similarity Graph Based on Shared Skills")
plt.axis("off")
plt.tight_layout()
plt.show()
```
